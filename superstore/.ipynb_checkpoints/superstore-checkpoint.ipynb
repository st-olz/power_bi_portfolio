# Superstore Sales

The goal of this project is to model and analyze data from a sales records database and then extract information for decision-making. Below are the questions we'll consider in this project.

- What were the total sales for the company?
- Which market generated the most sales on average?
- What were the profits by segment? Which segment has the most profit?

The dataset contains data on order details of customers for orders of a superstore in the U.S. It contains 13 columns:

- Order ID: all order IDs
- Customer ID: all customer information
- Customer Name: customer name
- Segment: segments of product
- Country: country of sales
- Market: continent/market of sales
- Product ID: list of product IDs
- Category: category of different product
- Sub-Category: sub-category of product
- Product Name: product name
- Sales: sales of different product
- Quantity: quantity of product sale
- Profit: profit made

## Data Cleaning
We need to clean the data before using it for modeling, data analysis, and creating visualizations. So, lets have a look at the data.

The current header isn't meaningful. The first row contains data, and the second row should be the header. So we have to "Use First Row as Headers" twice to correct it.

There is also a discrepancy between the columns' content and their data types. The format of the Date column values isn't consistent. It has two different delimiters. Some of them use the dd-mm-yyyy format, and others use dd/mm/yyyy. We have to replace all "/" by "-" to format all dates the same way.
